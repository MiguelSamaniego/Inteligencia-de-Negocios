```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Cargar datos 
archivo_excel = 'Columna.csv'
data = pd.read_csv(archivo_excel)
data
```

```python
replacements = {
    'é': 'e',
    'ñ': 'n',
    'í': 'i',
    'ó': 'o',
    'á': 'a',
    'ú': 'u',
    'ã³':'o',
    'ðÿ':' ',
    'š½ ':'si'
    # Añade más reemplazos según sea necesario
}

for original, replacement in replacements.items():
    data['text_punc_num'] = data['text_punc_num'].str.replace(original, replacement)
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

# Descargar las stopwords si es necesario
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

# Preprocesamiento del texto
def preprocesar(texto):

    if isinstance(texto, float):
        return []
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar caracteres no alfabéticos
    texto = re.sub(r'[^a-záéíóúñ]+', ' ', texto)
    # Tokenización
    palabras = word_tokenize(texto)
    # Eliminar stopwords
    palabras = [palabra for palabra in palabras if palabra not in stopwords.words('spanish')]
    # Lematización (opcional)
#     lemmatizer = WordNetLemmatizer()
#     palabras = [lemmatizer.lemmatize(palabra) for palabra in palabras]
    return palabras

# Ejemplo de uso
# texto_ejemplo = "Este es un ejemplo de texto para preprocesar."
# texto_procesado = preprocesar(texto_ejemplo)
# print(texto_procesado)
```

```python
data['texto_procesado'] = data['text_punc_num'].apply(preprocesar)
data['texto_procesado']
```

```python
data['texto_procesado'] = data['texto_procesado'].apply(lambda x: ' '.join(x))
```

```python
from gensim.models import Word2Vec

# Asumiendo que 'data' es tu DataFrame y 'texto_procesado' contiene los textos tokenizados
# Preparar los datos para Word2Vec (debe ser una lista de listas de palabras)
textos_tokenizados = [doc.split() for doc in data['texto_procesado']]

# Entrenar el modelo Word2Vec
modelo_w2v = Word2Vec(sentences=textos_tokenizados, vector_size=100, window=5, min_count=1, workers=4)

# Ejemplo: Obtener el vector de una palabra
vector_palabra = modelo_w2v.wv['ejemplo']
```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Palabras a visualizar
palabras = list(modelo_w2v.wv.index_to_key)[:100]  # Primeras 100 palabras
vectores = [modelo_w2v.wv[word] for word in palabras]

# Convertir 'vectores' en un array de NumPy
vectores_np = np.array(vectores)

# Reducción de dimensionalidad con t-SNE
tsne = TSNE(n_components=2, random_state=0)
vectores_2d = tsne.fit_transform(vectores_np)

# Crear un gráfico de dispersión
plt.figure(figsize=(15, 15))
for i, palabra in enumerate(palabras):
    plt.scatter(vectores_2d[i, 0], vectores_2d[i, 1])
    plt.text(vectores_2d[i, 0]+0.03, vectores_2d[i, 1]+0.03, palabra, fontsize=9)
plt.show()
```

En este tipo de visualización, las palabras que son semánticamente similares se colocan más cerca unas de otras en el espacio vectorial. Esto significa que si dos palabras o frases aparecen en contextos similares dentro del texto de entrenamiento, es probable que estén más cerca en el gráfico. Por ejemplo, palabras relacionadas con aspectos políticos podrían agruparse juntas, lo mismo que términos relacionados con economía, educación, etc.

Lo que muestra este gráfico es una representación bidimensional de datos multidimensionales. Para reducir la dimensionalidad y hacer que el gráfico sea comprensible, se suelen utilizar técnicas como PCA (Análisis de Componentes Principales) o t-SNE (t-distributed Stochastic Neighbor Embedding).

Cada punto en el gráfico representa una palabra, y la proximidad entre puntos sugiere una relación semántica más fuerte. Los colores de los puntos pueden representar diferentes categorías o grupos identificados dentro del espacio vectorial, aunque sin una leyenda o más contexto es difícil determinar qué representan exactamente.
