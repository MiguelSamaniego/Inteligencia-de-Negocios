```python
import pandas as pd

# Leer el archivo de Excel
df = pd.read_excel('DatosDelincu.xlsx')

# Reemplazar NaN con una cadena vacía y filtrar filas que empiezan con un número
df.iloc[:, 0] = df.iloc[:, 0].fillna('')
df = df[df.iloc[:, 0].str.match(r'^\d')]

# Dividir la primera columna en tres partes usando rsplit
df_dividido = df.iloc[:, 0].str.rsplit(',', n=2, expand=True)

# Dividir la primera parte de df_dividido en Index, Autor y Texto
df_dividido[['Index', 'Autor', 'Texto']] = df_dividido[0].str.split(',', n=2, expand=True)

# Asignar las dos últimas partes a Fecha de creacion y Hashtags
df_dividido['Fecha de creacion'] = df_dividido[1]
df_dividido['Hashtags'] = df_dividido[2]

# Eliminar las columnas originales y temporales
df_final = df_dividido.drop([0, 1, 2], axis=1)

# Mostrar las primeras filas del DataFrame final
print(df_final.head())
```

```python
nombre_archivo_nuevo = 'DatosProcesados.xlsx'
df_final.to_excel(nombre_archivo_nuevo, index=False)

print(f"DataFrame guardado como '{nombre_archivo_nuevo}'")
```
