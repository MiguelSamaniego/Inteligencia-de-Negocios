```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Cargar datos 
archivo_excel = 'Columna.csv'
data = pd.read_csv(archivo_excel)
data
```

```python
replacements = {
    'é': 'e',
    'ñ': 'n',
    'í': 'i',
    'ó': 'o',
    'á': 'a',
    'ú': 'u',
    'ã³':'o',
    'ðÿ':' ',
    'š½ ':'si'
    # Añade más reemplazos según sea necesario
}

for original, replacement in replacements.items():
    data['text_punc_num'] = data['text_punc_num'].str.replace(original, replacement)
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

# Descargar las stopwords si es necesario
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

# Preprocesamiento del texto
def preprocesar(texto):

    if isinstance(texto, float):
        return []
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar caracteres no alfabéticos
    texto = re.sub(r'[^a-záéíóúñ]+', ' ', texto)
    # Tokenización
    palabras = word_tokenize(texto)
    # Eliminar stopwords
    palabras = [palabra for palabra in palabras if palabra not in stopwords.words('spanish')]
    # Lematización (opcional)
#     lemmatizer = WordNetLemmatizer()
#     palabras = [lemmatizer.lemmatize(palabra) for palabra in palabras]
    return palabras

# Ejemplo de uso
# texto_ejemplo = "Este es un ejemplo de texto para preprocesar."
# texto_procesado = preprocesar(texto_ejemplo)
# print(texto_procesado)
```

```python
data['texto_procesado'] = data['text_punc_num'].apply(preprocesar)
data['texto_procesado']
```

```python
# Suponiendo que 'data' es tu DataFrame y 'texto_procesado' la columna con los textos
# ...

# Si los textos están en forma de listas de palabras, conviértelos a cadenas
data['texto_procesado'] = data['texto_procesado'].apply(lambda x: ' '.join(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2)
tfidf_matrix = tfidf_vectorizer.fit_transform(data['texto_procesado'])
```

```python
from sklearn.decomposition import TruncatedSVD

num_temas = 5  # Número de temas
lsa_model = TruncatedSVD(n_components=num_temas)
lsa_topic_matrix = lsa_model.fit_transform(tfidf_matrix)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Visualizar los documentos en el espacio de los temas
# Elegir dos temas para el gráfico
x = lsa_topic_matrix[:, 0]
y = lsa_topic_matrix[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.xlabel('Tema 1')
plt.ylabel('Tema 2')
plt.title('Distribución de los Documentos en los Temas')
plt.show()

# Mostrar las palabras más importantes para cada tema
terms = tfidf_vectorizer.get_feature_names_out()

for i, comp in enumerate(lsa_model.components_):
    terms_comp = zip(terms, comp)
    sorted_terms = sorted(terms_comp, key=lambda x: x[1], reverse=True)[:10]
    print("Tema " + str(i+1) + ": ")
    for t in sorted_terms:
        print(t[0])
    print(" ")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Selección de los dos primeros temas para la visualización
x = lsa_topic_matrix[:, 0]
y = lsa_topic_matrix[:, 1]

plt.figure(figsize=(12, 8))

scatter = plt.scatter(x, y, alpha=0.7, c=np.argmax(lsa_topic_matrix, axis=1), cmap='viridis')
plt.colorbar(scatter, label='Tema dominante')

plt.xlabel('Tema 1')
plt.ylabel('Tema 2')
plt.title('Distribución de los Documentos en los Temas')
plt.show()
```

```python
# Número de palabras clave para mostrar
num_palabras_clave = 10

# Mostrar las palabras más importantes para cada tema
for i, comp in enumerate(lsa_model.components_):
    terms_comp = zip(terms, comp)
    sorted_terms = sorted(terms_comp, key=lambda x: x[1], reverse=True)[:num_palabras_clave]
    
    # Palabras y pesos
    palabras = [term for term, _ in sorted_terms]
    pesos = [abs(weight) for _, weight in sorted_terms]
    
    plt.figure(figsize=(10, 6))
    y_pos = np.arange(len(palabras))
    
    plt.barh(y_pos, pesos, align='center', alpha=0.7)
    plt.yticks(y_pos, palabras)
    plt.gca().invert_yaxis()  # Invertir el eje Y para que las barras estén en orden descendente
    plt.xlabel('Importancia')
    plt.title(f'Palabras Clave del Tema {i + 1}')
    
    plt.show()
```

La primera imagen es el resultado de un modelo LDA, que ya he explicado en mi respuesta anterior.

Las otras tres imágenes parecen estar relacionadas con un modelo LSA. LSA es una técnica que se utiliza para analizar las relaciones entre una colección de documentos y los términos que contienen. A diferencia del LDA, que es probabilístico, LSA se basa en la descomposición de valores singulares (SVD, por sus siglas en inglés) para reducir la dimensión de los datos y descubrir las estructuras latentes (o "temas") en los datos textuales.

La segunda imagen muestra un gráfico de dispersión de la distribución de documentos en temas identificados por LSA. En el eje X, tienes el "Tema 1", y en el eje Y, el "Tema 2". Cada punto representa un documento, y su posición refleja la contribución o peso de esos temas en ese documento en particular. Por ejemplo, un punto ubicado cerca de (1,0) sería un documento que está fuertemente asociado con el Tema 1 y tiene poca o ninguna asociación con el Tema 2.

Las dos imágenes siguientes parecen listar las palabras clave asociadas con cada tema identificado por el modelo LSA. Cada tema tiene un conjunto de palabras que, según el modelo, capturan la esencia del tema. Estas palabras son elegidas en base a su contribución a la construcción de los conceptos subyacentes que LSA ha detectado en el conjunto de datos. Por ejemplo:

Tema 1: Incluye palabras como "charla", "feria", "palabra", lo que podría sugerir un tema relacionado con eventos sociales o conversaciones.
Tema 2: Contiene palabras como "generan", "pelea", "confundan", lo que podría indicar un tema relacionado con conflictos o desorden.
Tema 3: Incluye "clanes", "ligados", "corrupción", sugiriendo posiblemente un tema relacionado con la política o crimen organizado.
Tema 4 y Tema 5: No se puede observar completamente, pero parece continuar la lista de palabras relevantes para otros temas.
